```python
# prompt: Using dataframe df: scatter plot

import altair as alt

# Create a scatter plot of 'Units Sold' vs. 'Unit Price'
alt.Chart(df).mark_circle().encode(
    x='Units Sold',
    y='Unit Price',
    tooltip=['Product', 'Region', 'Total Sales'] # Add tooltips for more info
).interactive()
```

```python
plt.hist(df['Units Sold'],bins=20,color='blue')
plt.xlabel('Units Sold')
plt.ylabel('Frequency')
plt.title('Distribution of Unit Sold')
plt.show()
```
